## Algerian Forest Fires Dataset 
Data Set Information:

The dataset includes 244 instances that regroup a data of two regions of Algeria,namely the Bejaia region located in the northeast of Algeria and the Sidi Bel-abbes region located in the northwest of Algeria.

122 instances for each region.

The period from June 2012 to September 2012.
The dataset includes 11 attribues and 1 output attribue (class)
The 244 instances have been classified into fire(138 classes) and not fire (106 classes) classes.

#### Attribute Information:

1. Date : (DD/MM/YYYY) Day, month ('june' to 'september'), year (2012)
Weather data observations
2. Temp : temperature noon (temperature max) in Celsius degrees: 22 to 42
3. RH : Relative Humidity in %: 21 to 90
4. Ws :Wind speed in km/h: 6 to 29
5. Rain: total day in mm: 0 to 16.8
FWI Components
6. Fine Fuel Moisture Code (FFMC) index from the FWI system: 28.6 to 92.5
7. Duff Moisture Code (DMC) index from the FWI system: 1.1 to 65.9
8. Drought Code (DC) index from the FWI system: 7 to 220.4
9. Initial Spread Index (ISI) index from the FWI system: 0 to 18.5
10. Buildup Index (BUI) index from the FWI system: 1.1 to 68
11. Fire Weather Index (FWI) Index: 0 to 31.1
12. Classes: two classes, namely Fire and not Fire

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("Algerian_forest_fires_dataset_UPDATE.csv",header=1)
df.head()
```

## Data Cleaning

```python
# missing values
df.isnull().sum()
```

```python
df[df.isnull().any(axis=1)]
```

The dataset is converted into two sets based on region from 122th index, we can make a new column based on region

0: "Bejaia Region Dataset"

1: "Sidi-Bel Abbes Region Dataset"

Add new column with region

```python
df.loc[:122,'Region']=0
df.loc[122:,'Region']=1
df[['Region']] = df[['Region']].astype(int)
df.info()
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
# Removing teh null values
df = df.dropna().reset_index(drop=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df = df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
# Fix extra spaces in column names
df.columns=df.columns.str.strip()
df.columns
```

```python
## Change the required columns as interger datatype
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']]= df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

```python
## Change the other columns as float datatype
objects = [features for features in df.columns if df[features].dtype=='O']
objects
```

```python
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
```

```python
#Encoding of categorical data in Classes column
df['Classes']=np.where(df['Classes'].str.contains('not fire'),0,1)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
## Saving the cleaned dataset
df.to_csv("Algerian_forest_fires_cleaned_dataset_CLEANED.csv",index=False)
```

## Exploratory Data Analysis

```python
#Dropping day, month and year columns
df_copy = df.drop(['day','month','year'],axis=1)
df_copy.head()
```

```python
# Categories in Classes column
df_copy['Classes'].value_counts()
```

```python
# Plot density plot for all features
plt.style.use('seaborn-v0_8')
df_copy.hist(bins=50,figsize=(20,15))
plt.show()
```

```python
# Percentage of categories in Classes column for pie chart
df_copy['Classes'].value_counts(normalize=True)*100
```

```python
# Plotting pie chart
classlabels=['Fire','Not Fire']
plt.figure(figsize=(12,7))
percentage = df_copy['Classes'].value_counts(normalize=True)*100
plt.pie(percentage,labels=classlabels,autopct='%1.1f%%')
plt.title('Pie Chart of Classes')
plt.show()
```

```python
## Monthly fire analysis
dftemp = df.loc[df['Region']==1]
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='month',hue='Classes',data=dftemp)
plt.ylabel('Number of Fires', weight='bold')
plt.xlabel('Months',weight='bold')
plt.title('Fire analslysis og Bejaia Region by month',weight='bold')
plt.show()
```

```python
dftemp = df.loc[df['Region']==0]
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='month',hue='Classes',data=dftemp)
plt.ylabel('Number of Fires', weight='bold')
plt.xlabel('Months',weight='bold')
plt.title('Fire analslysis og Sidi-Bel Region by month',weight='bold')
plt.show()
```

From the above plots, It is observed that August month had the most number of forest fires in Sidi-Bel region and for Bejaia region, the most number for fires are in both July and August months.
